Here’s a structured documentation of the dataset fields for geographic aggregations (LSOA, MSOA, Ward, Borough), formatted for clarity and analysis:

---

### **Dataset Structure**
Comma-separated files with **202 columns**, aggregated by geographic area. Each file integrates food purchase data with census-derived metrics.

---

### **Core Fields**

#### **Identifier**
- `area_id`: Unique identifier for the geographic area [1][2].

---

### **Product Metrics**
- `weight`: Average food product weight (grams) [1].
- `volume`: Average drink product volume (liters) [1].
- `energy`: Nutritional energy per product (kcal) [1][2].
- `energy_density`: Calories per gram (kcal/g) [1].

---

### **Nutritional Composition**
**Nutrient weights (grams):**
- `carbs`, `sugar`, `fat`, `saturated fat`, `protein`, `fibre`  
  *(Note: Carbs include sugars; fats include saturated fats)* [1].

**Energy from nutrients (kcal):**  
- `energy_carbs`, `energy_sugar`, `energy_fat`, `energy_saturated fat`, `energy_protein`, `energy_fibre` [1].

---

### **Diversity Metrics**
- `h_nutrients_weight`: Entropy of nutrient weights [1].
- `h_nutrients_weight_norm`: Normalized nutrient weight entropy [0–1] [1].
- `h_nutrients_calories`: Entropy of energy from nutrients [1].
- `h_nutrients_calories_norm`: Normalized energy entropy [0–1] [1].

---

### **Category Fractions**
**Product type distribution:**
- `f_{category}`: Fraction of products in category (e.g., `f_dairy`, `f_fish`).  
  *Categories*: `beer`, `dairy`, `eggs`, `fats & oils`, `fish`, `fruit & veg`, `grains`, `red meat`, `poultry`, `readymade`, `sauces`, `soft drinks`, `spirits`, `sweets`, `tea & coffee`, `water`, `wine` [1][2].

**Weight distribution by category:**
- `f_{category}_weight`: Fraction of total weight from category (e.g., `f_dairy_weight`).

**Category diversity metrics:**
- `h_category`: Entropy of product categories [1].
- `h_category_norm`: Normalized category entropy [0–1].
- `h_category_weight`: Entropy of category weights.
- `h_category_weight_norm`: Normalized category weight entropy [0–1].

---

### **Census Integration**
- `representativeness_norm`: Normalized ratio of unique customers to census population [0–1] [1][2].
- `population`: Total residents (2015 census) [1][2].
- `male`, `female`: Gender distribution [1].
- `age_0_17`, `age_18_64`, `age_65+`: Age groups [1][2].
- `avg_age`: Average resident age [1].
- `area_sq_km`: Surface area (km²) [2][26].
- `people_per_sq_km`: Population density [1][2].

---

### **Transaction Metadata**
- `transaction_days`: Unique dates with purchases [1].
- `num_transactions`: Total products purchased by Clubcard residents [1][2].
- `man_day`: Cumulative customer purchase days [1].

---

### **Statistical Measures**
For applicable fields (e.g., `energy`, `weight`):
- `{field}_std`: Standard deviation.
- `{field}_ci95`: 95% confidence interval for the mean.
- `{field}_perc{2.5,25,50,75,97.5}`: Percentile values.

---

This structure enables granular analysis of food purchasing patterns, nutrient intake, and demographic correlations across geographic areas. The integration of census data allows for cross-referencing with socioeconomic indicators [1][2][21].

## Concatenando todos los DataSets

```python
meses = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec"]
area_id_to_name = {
    "E09000001": "City of London",a
    "E09000002": "Barking and Dagenham",a
    "E09000003": "Barnet",a
    "E09000004": "Bexley",a
    "E09000005": "Brent",a
    "E09000006": "Bromley",a
    "E09000007": "Camden",a
    "E09000008": "Croydon",a
    "E09000009": "Ealing",a
    "E09000010": "Enfield",a
    "E09000011": "Greenwich",a
    "E09000012": "Hackney",a
    "E09000013": "Hammersmith and Fulham",a
    "E09000014": "Haringey",a
    "E09000015": "Harrow",a
    "E09000016": "Havering",a
    "E09000017": "Hillingdon",a
    "E09000018": "Hounslow",a
    "E09000019": "Islington",a
    "E09000020": "Kensington and Chelsea",a
    "E09000021": "Kingston upon Thames",a
    "E09000022": "Lambeth",a
    "E09000023": "Lewisham",a
    "E09000024": "Merton",a
    "E09000025": "Newham",a
    "E09000026": "Redbridge",a
    "E09000027": "Richmond upon Thames",a
    "E09000028": "Southwark",a
    "E09000029": "Sutton",a
    "E09000030": "Tower Hamlets",a
    "E09000031": "Waltham Forest",a
    "E09000032": "Wandsworth",a
    "E09000033": "Westminster"a
}
dataframes = []
for mes in meses:
    nombre_archivo = f"./Datasets/{mes}.csv"
    df = pd.read_csv(nombre_archivo)
    df["mes"] = mes
    df["area_name"] = df["area_id"].map(area_id_to_name)
    dataframes.append(df)
fusion_df=pd.concat(dataframes,ignore_index=True)
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
display(fusion_df)
```

```python
media = fusion_df.groupby(["area_id","area_name"])["weight"].mean().reset_index()
plt.figure(figsize=(15,10))
plt.bar(media["area_name"],media["weight"])
plt.xticks(rotation=90)
plt.xlabel('Boroughs de Londres')
plt.ylabel('Media de weight')
plt.title('Media por Borough de Londres (ordenado por ID)')
plt.show()

```

```python
media = fusion_df.groupby(["area_id","area_name","mes"])["weight"].mean().reset_index()
plt.figure(figsize=(20,10))
media.pivot(index="area_name",columns="mes",values="weight").plot(kind="bar",stacked=False)
plt.title('Media de Valores por Borough y Mes', fontsize=16)
plt.xlabel('Boroughs de Londres', fontsize=12)
plt.ylabel('Valor Medio', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Mes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
